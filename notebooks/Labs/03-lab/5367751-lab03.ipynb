```python
# Packages yippiee

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
import cv2
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import SGDClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
from sklearn.metrics import classification_report
from sklearn.utils import shuffle
import seaborn as sns
import random
import glob
from PIL import Image
from collections import Counter
```

```python
# Step 2: Load the dataset
data_dir = 'data'

image_paths = glob.glob(os.path.join(data_dir, "*jpg"))

# print("Sample files:")
# for path in image_paths[:5]:
#     print(os.path.basename(path))


def extract_label_from_filename(filename):
    basename = os.path.basename(filename)
    parts = basename.split('_')
    if len(parts) >= 4:
        try:
            label = int(parts[3].split('.')[0])
            return label
        except (ValueError, IndexError):
            print("Error")

    return None

labels = [extract_label_from_filename(path) for path in image_paths]
unique_labels = sorted(set([l for l in labels if l is not None]))
print(f"Total number of image: {len(labels)}")
print(f"Number of unique classes: {len(unique_labels)}")

class_distribution = Counter(labels)
print("Class distribution")
for label, count in sorted(class_distribution.items()):
    if label is not None:
        print(f" Class {label}: {count} images")

if image_paths:
    sample_img = Image.open(image_paths[0])
    print(f"Image size: {sample_img.size}")

# Function to show sample images from each class
def show_sample_img_by_class(image_paths, labels, unique_labels, sample_per_class=3):
    # mapping from labels to img paths
    label_to_paths = {}
    for path, label in zip(image_paths, labels):
        if label not in label_to_paths:
            label_to_paths[label] = []
        label_to_paths[label].append(path)

    # grid ddimensions
    num_classes = len(unique_labels)
    cols = 5
    rows = (num_classes + cols -1) // cols

    plt.figure(figsize=(15, 3*rows))

    for i, label in enumerate(unique_labels):
        if label in label_to_paths:
            sample_paths = random.sample(
                label_to_paths[label],
                min(sample_per_class, len(label_to_paths[label]))
            )

            for j, path in enumerate(sample_paths[:1]):
                img = Image.open(path)
                plt.subplot(rows, cols, i + 1)
                plt.imshow(img, cmap='gray')
                plt.title(f"Class: {label}")
                plt.axis('off')


    plt.tight_layout()
    plt.show()

show_sample_img_by_class(image_paths, labels, unique_labels)
```

```python
# step 3: Splt and sample the dataset

# Function to perform stratified t/t split
def create_stratified_splits(image_paths, labels, train_size=5000, test_size=1000):
    paths = np.array(image_paths)
    y = np.array(labels)

    train_paths, test_paths, train_labels, test_labels = train_test_split(
        paths,
        y,
        train_size=train_size,
        test_size=test_size,
        stratify=y,
        random_state=42
    )

    train_dist = Counter(train_labels)
    test_dist = Counter(test_labels)

    print(f"Training set: {len(train_paths)} images")
    print(f"Testing set: {len(test_paths)} images")

    print("\nClass distribution in training set:")
    for label, count in sorted(train_dist.items()):
        print(f"  Class {label}: {count} images ({count/len(train_labels)*100:.1f}%)")

    print("\nClass distribution in testing set:")
    for label, count in sorted(test_dist.items()):
        print(f"  Class {label}: {count} images ({count/len(test_labels)*100:.1f}%)")

    return train_paths, test_paths, train_labels, test_labels

train_paths_5k, test_paths, train_labels_5k, test_labels = create_stratified_splits(image_paths, labels)

# Loads and processes images for ML classifiers
# Implements step 4: perform necessary data reshaping
def load_images(image_paths, labels):
    images = []
    v_labels = []

    for path, label in zip(image_paths, labels):
        try:
            img = cv2.imread(path, cv2.IMREAD_GRAYSCALE)
            if img is None:
                print(f"Image did not lead {path}")
                continue

            img_flattened = img.flatten()
            
            images.append(img_flattened)
            v_labels.append(label)
        except Exception as e:
            print(e)


    X = np.array(images)
    y = np.array(v_labels)

    return X, y

X_train_5k, y_train_5k = load_images(train_paths_5k, train_labels_5k)
X_test, y_test = load_images(test_paths, test_labels)

print(f"Loaded training data shape: {X_train_5k.shape}")
print(f"Loaded testing data shape: {X_test.shape}")
```

```python
# Step 5: Initialise the classifier

knn_classifier = KNeighborsClassifier(n_neighbors=3)
dt_classifier = DecisionTreeClassifier()
sgd_classifier = SGDClassifier(max_iter=250, random_state=42)
```

```python
# Step 6: Fit the classifier to the training data
print("Training KNN")
knn_classifier.fit(X_train_5k, y_train_5k)
dt_classifier.fit(X_train_5k, y_train_5k)
sgd_classifier.fit(X_train_5k, y_train_5k)

print("All classifiers are trained")
```

```python
# Step 7: Evaluate the trained model on the testing data
knn_predictions = knn_classifier.predict(X_test)
print("KNN prediction completed")

dt_predictions = dt_classifier.predict(X_test)
print("DT prediction completed")

sgd_predictiosn = sgd_classifier.predict(X_test)
print("SGD prediction completed")
```

```python
# Step 8: Report the performance of each classifier

# Reusbal function to display confusion matrix
def plot_confusion_matrix(cm, class_names, title):
    plt.figure(figsize=(10, 8))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=class_names, yticklabels=class_names)
    plt.xlabel('Predicted Label')
    plt.ylabel('True Label')
    plt.title(title)
    plt.tight_layout()
    plt.show()

def report_classifier_performance(y_true, y_pred, classifier_name):
    acc = accuracy_score(y_true, y_pred)
    prec = precision_score(y_true, y_pred, average='weighted')
    rec = recall_score(y_true, y_pred, average='weighted')
    f1 = f1_score(y_true, y_pred, average='weighted')


    print(f"\n{classifier_name} Performacne Metrics")
    print(f"Accuracy: {acc:.4f}")
    print(f"Precision: {prec:.4f}")
    print(f"Recall: {rec:.4f}")
    print(f"F1-Score: {f1:.4f}")

    cm = confusion_matrix(y_true, y_pred)
    plot_confusion_matrix(cm, unique_labels, f"{classifier_name} Confusion Matrix")


    return {
        'classifier': classifier_name,
        'accuracy': acc,
        'precision': prec,
        'recall': rec,
        'f1': f1
    }

print("\nPerformance metrics for 5k training dataset:")
knn_5k_metrics = report_classifier_performance(y_test, knn_predictions, "KNN 5K")
dt_5k_metrics = report_classifier_performance(y_test, dt_predictions, "DT 5K")
sgd_5k_metrics = report_classifier_performance(y_test, sgd_predictiosn, "SGD 5K")

results_5k = [knn_5k_metrics, dt_5k_metrics, sgd_5k_metrics]
```

```python
# For the 10k dataset

print("\nTraining classifiers on 10k")

# Creating the larger training set (10k)
train_paths_10k, test_paths, train_labels_10k, test_labels = create_stratified_splits(image_paths, labels, train_size=10000, test_size=1000)

X_train_10k, y_train_10k = load_images(train_paths_10k, train_labels_10k)
X_test, y_test = load_images(test_paths, test_labels)

print(f"Loaded training data shape: {X_train_10k.shape}")
print(f"Loaded testing data shape: {X_test.shape}")

# Step 5: Initialise the classifier
knn_classifier_10k = KNeighborsClassifier(n_neighbors=3)
dt_classifier_10k = DecisionTreeClassifier()
sgd_classifier_10k = SGDClassifier(max_iter=250, random_state=42)

# Step 6: Fit the classifier to the training data
print("Training KNN")
knn_classifier_10k.fit(X_train_10k, y_train_10k)
dt_classifier_10k.fit(X_train_10k, y_train_10k)
sgd_classifier_10k.fit(X_train_10k, y_train_10k)

print("All classifiers are trained")

# Step 7: Evaluate the trained model on the testing data
knn_predictions_10k = knn_classifier_10k.predict(X_test)
print("KNN prediction completed")

dt_predictions_10k = dt_classifier_10k.predict(X_test)
print("DT prediction completed")

sgd_predictiosn_10k = sgd_classifier_10k.predict(X_test)
print("SGD prediction completed")

print("\nPerformance metrics for 5k training dataset:")
knn_10k_metrics = report_classifier_performance(y_test, knn_predictions_10k, "KNN 10K")
dt_10k_metrics = report_classifier_performance(y_test, dt_predictions_10k, "DT 10K")
sgd_10k_metrics = report_classifier_performance(y_test, sgd_predictiosn_10k, "SGD 10K")

results_10k = [knn_10k_metrics, dt_10k_metrics, sgd_10k_metrics]
```

```python
# Comparing performance

'''
Comparing the performance between 5k and 10k training datasets, it seems that the classifiers benefitted 
with increased training data, as the accuracy among all the classifiers saw an increase.
'''

print("\nComparing Performance of 5k vs 10k training datasets:")
for i, c in enumerate(["KNN", "DT", "SGD"]):
    print(f"\n{c}:")
    print(f"  Accuracy: {results_5k[i]['accuracy']:.4f} -> {results_10k[i]['accuracy']:.4f}")
    print(f"  Precision: {results_5k[i]['precision']:.4f} -> {results_10k[i]['precision']:.4f}")
    print(f"  Recall: {results_5k[i]['recall']:.4f} -> {results_10k[i]['recall']:.4f}")
    print(f"  F1-Score: {results_5k[i]['f1']:.4f} -> {results_10k[i]['f1']:.4f}")

```
